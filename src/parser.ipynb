### USGS parser

```python
outfile = open('catalog_usgs.csv', 'ab')

first = True

for year in range(2010, 2020):

    months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

    for i in range(0, 12):

        USGS_query_url = f"{USGS_base_url}"\
                        f"starttime={year}-{months[i]}-01&"\
                        f"endtime={year if i != 11 else year+1}-{months[i+1] if i != 11 else '01'}-01&"\
                        f"orderby=time-asc"
                        
        urllib.request.urlretrieve(USGS_query_url, filename='bull_tmp')

        with open('bull_tmp', 'rb') as csvfile:

            if not first:
                csvfile.readline()
            
            fileContent = csvfile.read()

            outfile.write(fileContent)

        first = False


    print(f"{year}-{year+1} got")


outfile.close()
```
